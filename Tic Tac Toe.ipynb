```python
# board is a list of strings representing the board
board = ["-", "-", "-", "-", "-", "-", "-", "-", "-"]

# welcome the 2 players to play the game 
print("Welcome to Tic Tac Toe!")

def display_board():
    # function to display the game board   
    # a 3x3 grid for the 2 players to play the game and a display for the 2 players to know what numbers to input
    
    print(board[0] + " | " + board[1] + " | " + board[2] + "     1 | 2 | 3")
    print("----------") 
    print(board[3] + " | " + board[4] + " | " + board[5] + "     4 | 5 | 6")
    print("----------") 
    print(board[6] + " | " + board[7] + " | " + board[8] + "     7 | 8 | 9")
    
def play_a_turn(mark):
    # function to play a single game of tic tac toe 
    
    while True:  
        # capturing a player's input
        # try except block to ensure that the player keys in only integer value  
        # ask for the player's input again if he/she keys in non-integer value   
        try:
            spot = int(input(f"Choose a spot({mark}): ")) - 1
        except:
            print("Please enter a valid input!")
            continue
            
        # if loop to check that the the player's input is between 1 to 9
        # so that the spot on the board is between 0 and 8 and that the spot is not occupied  
        # if the condition is met, leave a mark on the board, break out from the while-loop
        # if the condition is not met, repeat the while-loop 
        # ask for the player's input again until his/her input is between 1 to 9
        if 0 <= spot <= 8 and board[spot] == "-":
            board[spot] = mark
            break
        else:
            print("Wrong input! Please try again!") 
            
def check_winner(mark):
    # function to check for a winning combination
    # total of 8 winning combinations
    # if loop to check for a winning combination
    # returns True if there is a winning combination, declaring that a player has won
    # returns False if there is no winning combination, declaring that nobody has won
    
    # check the rows for a win
    if board[0] == board[1] == board[2] == mark:
        print(f"Player {mark} won")
        return True
    if board[3] == board[4] == board[5] == mark:
        print(f"Player {mark} won")
        return True
    if board[6] == board[7] == board[8] == mark:
        print(f"Player {mark} won")
        return True
    
    # check the columns for a win
    if board[0] == board[3] == board[6] == mark:
        print(f"Player {mark} won")
        return True
    if board[1] == board[4] == board[7] == mark:
        print(f"Player {mark} won")
        return True
    if board[2] == board[5] == board[8] == mark:
        print(f"Player {mark} won")
        return True
    
    # check the diagonals for a win
    if board[0] == board[4] == board[8] == mark:
        print(f"Player {mark} won")
        return True
    if board[6] == board[4] == board[2] == mark:
        print(f"Player {mark} won")
        return True
    
    else:
        print("Nobody has won")
        return False
    
def game():
    # function to play game 
    # for loop since the max number of turns is 9 
    # if loop to switch between player x and player o 
    # if odd turns, player x to play 
    # if even turns, player o to play
    
    for counter in range(1,10):
        display_board()       
        if counter %2 == 0:
            mark = "o"
        else:
            mark = "x"       
        play_a_turn(mark)
        
        # check for a winner 
        # returns True if there is a winner
        # breaks and ends the game
        check_outcomes = check_winner(mark)
        if check_outcomes == True:
            break
    
    # display the final result on the board            
    display_board()
    
game()
```
