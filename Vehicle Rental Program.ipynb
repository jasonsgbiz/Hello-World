```python
class Vehicle:
    # constructor to initialize the instance attributes      
    def __init__(self, mpg, vin):
        
        self._mpg = mpg
        self._vin = vin
        self._reserved = False

    # instance method to get type of vehicle 
    def getType(self):
        
        if isinstance(self, Car):
            return "Car"
        elif isinstance(self, Van):
            return "Van"
        elif isinstance(self, Truck):
            return "Truck"

    # instance method to get vin of vehicle
    def getVin(self):
                
        return self._vin

    # instance method to get description of vehicle
    def getDescription(self):
                
        return "mpg: {}   vin: {}".format(self._mpg, self._vin)
    
    # instance method to check whether vehicle is reserved
    def isReserved(self):
                
        return self._reserved

    # instance method to set the vehicle to reserved
    def setReserved(self, reserved):
        
        self._reserved = reserved
    
#    @property
#    def reserved(self):
#        return self._reserved

#    @reserved.setter
#    def reserved(self, status):
#        if isinstance(status,bool):
#            self._reserved = status
            
class Car(Vehicle):
    # constructor to initialize the instance attributes    
    # the Car (child) class inherits the attributes and methods of the Vehicle (parent) class
    def __init__(self, make_model, mpg, num_passengers, num_doors, vin):
 
        super().__init__(mpg, vin)

        self._make_model = make_model
        self._num_passengers = num_passengers
        self._num_doors = num_doors

    # instance method to get description of car
    def getDescription(self):
 
        return "make_model: {}   mpg: {}   passengers: {}   doors: {}   vin: {}".format(self._make_model,\
            self._mpg, self._num_passengers, self._num_doors, self._vin) 

class Van(Vehicle):
    # constructor to initialize the instance attributes    
    # the Van (child) class inherits the attributes and methods of the Vehicle (parent) class
    def __init__(self, make_model, mpg, num_passengers, vin):
         
        super().__init__(mpg, vin)

        self._make_model = make_model
        self._num_passengers = num_passengers

    # instance method to get description of van
    def getDescription(self):
 
        return "make_model: {}   mpg: {}   passengers: {}   vin: {}".format(self._make_model,\
            self._mpg, self._num_passengers, self._vin) 

class Truck(Vehicle):
    # constructor to initialize the instance attributes    
    # the Van (child) class inherits the attributes and methods of the Vehicle (parent) class
    def __init__(self, length, mpg, num_rooms, vin):

        super().__init__(mpg, vin)
        
        self._length = length
        self._num_rooms = num_rooms

    # instance method to get description of truck
    def getDescription(self):
 
        return "length: {}   mpg: {}   rooms: {}   vin: {}".format(self._length,\
            self._mpg, self._num_rooms, self._vin)
```
